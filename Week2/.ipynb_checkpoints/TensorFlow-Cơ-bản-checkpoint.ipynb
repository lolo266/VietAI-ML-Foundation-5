###  Placeholder & feed dict 
Giúp đưa dữ liệu đầu vào(input) vào chương trình

```python
import tensorflow as tf
input_placeholder = tf.placeholder(tf.int32)
node_three = tf.constant(3)
sum_node = input_placeholder + node_three ## equivalent to tf.add(two_node, three_node)
sess = tf.Session()
print(sess.run(sum_node, feed_dict={input_placeholder: 2}))
```

## Viết chương trình theo Tensorflow

Cài đặt chương trình theo hàm cost

Tìm giá trị nhỏ nhất của hàm số sau:

J(w) = w^2 - 10w + 25

Thực hiện tuần tự các bước sau:

1. Khởi tạo biến
2. Bắt đầu session
3. Giải thuật
4. Close session

```python
coefficients = np.array([[1.0], [-10], [25.]]) #Định nghĩa các hệ số
#Truyền gía trị vào
w = tf.Variable(initial_value=0, dtype=tf.float32)
x = tf.placeholder(dtype=tf.float32, shape=(3,1))

cost = w**2 * x[0] + w*x[1] + x[2]
#tìm điểm cực tiểu dựa trên giải thuật GD
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
init = tf.global_variables_initializer()
session = tf.Session()
session.run(init)

#chạy vòng lặp để tìm giá trị cực tiểu của hàm số
num_iterations = 1000
for _ in range(num_iterations):
  session.run(train, feed_dict={x: coefficients})

print(session.run(w))
```
